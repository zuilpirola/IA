## 1. Recursive

Consider the DFS (Depth First Search) code below:



```python
g = {
    'A': ['B', 'C', 'E'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': ['D', 'F'],
    'F': ['C'],
    'G': ['F'],
    'H': []
}

def dfs(neighbour_list, root_node, goal):

    #nós ja visitados
    visited_nodes = set()

    # lista iniciada a partir do nó inicial
    q = [root_node]

    #enquanto q existir e a flag found for false
    while q:

        #exclui-se o ultimo elemento da q
        node = q.pop(0)

        #print do visiting node
        print('visiting node', node)

        #se o nó analisado for o nó buscado
        if node == goal:
            #print para avisar que achou o nó
            print('found goal node')
            break

        #caso o nó analisado nao seja o nó buscado
        elif node not in visited_nodes:
            #adicionar o nó analisado na lista de nós visitados
            visited_nodes.add(node)

            q = sorted(set(neighbour_list[node]) - visited_nodes) + q

            print('q is', q)

    return list(visited_nodes)

print(dfs(g, 'A', 'G'))
```

Rewrite the code above making it a recursive function (without using `while`).

```python
def dfs(neighbour_list, root_node, goal, visited_nodes=None, stack=None):
    # Verifica se visited_nodes e stack são None, e se for, inicia-os.
    if visited_nodes is None:
        visited_nodes = set()
    if stack is None:
        stack = [root_node]

    # Se a stack estiver vazia, retorna visited_nodes.
    if not stack:
        return list(visited_nodes)

    # Remove o último elemento da stack.
    node = stack.pop(0)

    # Imprime o nó visitado.
    print('visiting node', node)

    # Se o nó for o objetivo, imprime que o objetivo foi encontrado e retorna visited_nodes.
    if node == goal:
        print('found goal node')
        return list(visited_nodes)

    # Adiciona o nó visitado a visited_nodes.
    visited_nodes.add(node)

    # Adiciona os filhos do nó na frente da stack.
    stack = sorted(set(neighbour_list[node]) - visited_nodes) + stack


    print('stack is', stack)

    # Chama a função novamente recursivamente.
    return dfs(neighbour_list, root_node, goal, visited_nodes, stack)
```

```python
print(dfs(g, 'A', 'G'))
```

## 3. Working with queue

In the codes above (DFS), stacks are used as basic structures, although BFS uses queues. Write two functions below that work with queues:

- One that add data into a queue;
- One that remove and returns the data from it.


Tip: use queue objects from library collections.

```python
from collections import deque

def enqueue(queue, item):
    """
    Adds data into a queue.

    Parameters:
    - queue: The queue to which the data will be added.
    - item: The data to be added into the queue.
    """
    queue.append(item)

def dequeue(queue):
    """
    Removes and returns the data from a queue.

    Parameters:
    - queue: The queue from which data will be removed.

    Returns:
    - The data removed from the queue.
    """
    return queue.popleft()
```

```python
# Create a queue
my_queue = deque()

# Add data into the queue
enqueue(my_queue, 1)
enqueue(my_queue, 2)
enqueue(my_queue, 3)
print(my_queue)

# Remove and print data from the queue
print(dequeue(my_queue))  # Output: 1
print(dequeue(my_queue))  # Output: 2
print(my_queue)
print(dequeue(my_queue))  # Output: 3
```
